```python
import transformers
import datasets
import numpy
```

```python
dset=datasets.load_dataset("json", data_files={"fi":"http://dl.turkunlp.org/TKO_8964_2023/news-fi-2019.jsonl"})
print(dset)
dset["fi"] = dset["fi"].shuffle(seed=42).select(range(1000))
```

```python
dset["fi"]["text"][:5]
```

```python
feature_pipeline = transformers.pipeline(task="feature-extraction",model="sentence-transformers/paraphrase-xlm-r-multilingual-v1", return_tensors=True,device=0) 
# unnecessary to have: tokenizer="sentence-transformers/paraphrase-xlm-r-multilingual-v1",
```

```python
fi_embedded = feature_pipeline(transformers.pipelines.pt_utils.KeyDataset(dset["fi"], "text"), batch_size=64, truncation="only_first")

import torch 
fi_embedded_pooled = [torch.mean(elem,axis=1) for elem in fi_embedded]
all_fi = torch.vstack(fi_embedded_pooled)
```

```python
def question_embeddings(questions):
  q_embedded = feature_pipeline(questions, truncation="only_first")
  q_embedded_pooled = [torch.mean(elem,axis=1) for elem in q_embedded]
  all_q = torch.vstack(q_embedded_pooled)
  return all_q
```

```python
def get_pairs(M):
  aligned=numpy.argsort(-M,axis=-1)
  pairs=[]
  #for i in range(M.shape[0]): #M.shape[0] is the number of rows / input documents # UNNECESSARY TO HAVE THIS LOOP SINCE I WILL BE LOOKING ONLY AT ONE QUESTION AT A TIME
  for i in range(10): # M.shape[1]
    j=aligned[0,i] # by using i we take ten of the best ones
    score=M[0,j]
    pairs.append((j,score))
    #print(score)
  #pairs.sort(key=lambda ijs:ijs[1]) #Sort the matches in descending order of their similarity -> already sorted(?)
  return pairs
```

```python
def get_similar(dset, pairs):
  similar_texts = []
  for j,score in pairs[:10]: # change to some number to get top k, descending order so these are the best
      print(score)
      print(dset["fi"]["text"][j])
      print()
      similar_texts.append(dset["fi"]["text"][j])
  return similar_texts
```

### qa model in use

From here on use the qa model on each of the top k texts to get the answer that is most likely (so look at score)

```python
model = "TurkuNLP/bert-base-finnish-cased-squad2"

tokenizer = "TurkuNLP/bert-base-finnish-cased-v1"

qa_pipeline = transformers.pipeline(model=model, tokenizer=tokenizer, device=0) # task is "question-answering" but is unnecessary to put here
```

```python
# the code that helps with trainer.predict()
#https://github.com/huggingface/notebooks/blob/main/examples/question_answering.ipynb 
#https://huggingface.co/course/chapter7/7#fine-tuning-the-model 
```

```python
def get_answer(similar_texts, question):
  results = []
  for one in similar_texts: # similar texts could be turned into a dataset and then this pipeline wouldn't have to be called many times?
    res = qa_pipeline(question=question, context=one, top_k=5) # doc_stride is a thing that is on by default at 128, so will work out of the box with long contexts!!
    # removed handle_impossible_answer=True to get something other than emptiness at the top
    results.append(res)
  # sort by the biggest score (probability)
  flat_list = [item for sublist in results for item in sublist]
  newlist = sorted(flat_list, key=lambda d: d['score'], reverse=True) 
  pprint(newlist[:5])
```

```python
from pprint import pprint
# this is the main code to get similar stuff
import sklearn.metrics

for i in range(len(questions)): # IMPORTANT this should be changed to include the questions, loop i in range the questions and use all_q[i] in here?
  one = question_embeddings([questions[i]])
  M=sklearn.metrics.pairwise.cosine_similarity(one.numpy(),all_fi.numpy())
  pairs = get_pairs(M)
  print(questions[i])
  similar_texts = get_similar(dset, pairs)
  print(questions[i])
  #pprint(similar_texts[0][:500])
  get_answer(similar_texts, questions[i])
```
